a. the average of the numbers (two decimal places suggested)

d. the median of the numbers. Note that if there was 2 numbers that qualify

    Enter third number: 1

`Functions`:
+ question2():
    - This is main function.
    - Takes input from user to accept 3 integers or 4 integers.
    - Validate input using valid_input_message function.
    
+ find_max and find_min:
    - get max and min value from the input.
    
+ sort 
    - returns the sorted list 
    
+ median 
    - calculates the median

```python
def valid_input_message(msg):
    '''
    msg: string input to display to the user. 
        Function trys to return the first valid inout given by the user.
        and retry if invalid inputs are given.
    '''
    i = input(msg)
    while(True):
        try:
            i = int(i)
            return i
        except:
            print("You Entered Non-Integer Value.")
            i = input(msg)

def median(l):
    '''
    l: list input return the median of the list.
        If the length of list is 
    '''
    if len(l)%2==1:
        return l[len(l)//2]
    else:
        return (l[len(l)//2]+l[(len(l)//2)-1])/2
    
def find_max(num):
    '''
    num:list returns the maximum value from the list.
    '''
    a = None
    for i in num:
        if not(a):
            a = i
        else:
            if a<i:
                a = i
    return a

def find_min(num):
    '''
    num: return the minimum value from the list.
    '''
    a = None
    for i in num:
        if not(a):
            a = i
        else:
            if a>i:
                a = i
    return a

def sort_list(num):
    '''
    num: List return the sorted list.
    Logic: Remove the minimum from list and append in second list.
    returns the second list.
    '''
    empty_list = []
    while num:
        nn = find_min(num)
        empty_list.append(nn)
        num.remove(nn)
    return empty_list

def find_avg(num):
    '''
    num: List returns the avg of the num
    Logic: sum(numbers)/len(numbers)
    '''
    sm=0
    for k,i in enumerate(num,1):
        sm+=i
    return sm/k
def question2():
    '''
    ***********MAIN FUNCTION********
    - Accepts: User Input to take 3 or 4 integers.
    - Accepts: User Input for the Integers.
    - calls avg, find_max, find_min, sort, median to find the values.
    - 
    '''
    while(True):
        ints = valid_input_message("Enter number of integers 3 or 4: ")
        if 2<ints<5:
            break
    
    numbers = []
    for i in range(ints):
        numbers.append(valid_input_message("Enter number {}: ".format(i+1)))

    avg = find_avg(numbers)
    max_num = find_max(numbers)
    min_num = find_min(numbers)
    med = median(sorted(numbers))
    
    print("Average of numbers = {} ".format(avg))
    print("Maximum of entered numbers = {}".format(max_num))
    print("Minimum of entered numbers = {}".format(min_num))
    print("Median of entered numbers = {}".format(med))
question2()
```
